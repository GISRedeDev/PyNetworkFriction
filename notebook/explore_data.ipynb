```python
import pandas as pd

df = pd.read_csv("../data/routes_incidents.csv")
df
```

```python
df.columns
```

```python
# Drop the 'fatalities' column
df = df.drop(columns='fatalities_sum')

df_grouped = df.groupby(['from_pcode', 'to_pcode', 'euclidean_dist', 'shortest_path_length', 'event_date'])['disorder_type'].value_counts().unstack(fill_value=0).reset_index()
```

```python
df_grouped.to_csv("GROUPED.csv", index=False)
```
